```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= ' unicode_escape')
```

```python
df.shape
```

```python
df.head(11)
```

```python
df.info()
```

```python
df.drop(columns=['Status', 'unnamed1'], inplace=True, errors='ignore')
```

```python
df.info()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace= True)
```

```python
df.shape
```

```python
#change data type
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#check column name 
df.columns
```

```python
# mathematically operation summary
df.describe()
```

```python
sns.countplot(x= 'Gender', data=df,)
```

```python
df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Gender', y='Amount',data = sales_gen)
```

```python
### Gender-wise Purchase Analysis
This chart shows how purchase amounts vary between male and female customers.
```

```python
# Age group analysis
```

```python
ax=sns.countplot(data=df,x='Age Group', hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs Age group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Age Group', y= 'Amount', data = sales_age)
```

```python
### Gender-wise Purchase Analysis
#This chart shows how purchase amounts vary between Age Group 
```

```python
#State
```

```python
# State vs Order
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_state,x='State', y= 'Orders')
```

```python
# amount vs state 
```

```python
# Amount vs Order
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_state,x='State', y= 'Amount')
```

```python
# Marital status
```

```python
ax= sns.countplot(data= df, x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state,x='Marital_Status', y= 'Amount',hue='Gender')
```

```python
#Occupation
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data= df, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# occupation vs sale/amount
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state,x='Occupation', y= 'Amount')
```

```python
sns.set(rc={'figure.figsize':(25,10)})
ax=sns.countplot(data= df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state,x='Product_Category', y= 'Amount')
```
